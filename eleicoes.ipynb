```python
import basedosdados as bd

# Para carregar o dado direto no pandas
df = bd.read_table(dataset_id='br_poder360_pesquisas',
table_id='microdados',
billing_project_id="pesquisas-eleitorais-375014")
```

```python
df
```

```python
#Identificando as colunas do dataframe para filtrarmos depois somente para ficarmos com as necessárias 
for coluna_nome in df.columns: 
    print(coluna_nome)
```

```python
#Criando listas para identificarmos os anos,cargos,presidentes,partidos,institutos e UFs
cargos = []
institutos = []
candidatos = []
partidos = []
ufs = []
tipos = []
cenarios = []
ids = []
#Cargos
for cargo in df['cargo']:
    if cargo not in cargos:
        cargos.append(cargo)
print(cargos)
#Institutos
for instituto in df['instituto']:
    if instituto not in institutos:
        institutos.append(instituto)
print(institutos)
#Nome do Candidato
for nome in df['nome_candidato']:
    if nome not in candidatos:
        candidatos.append(nome)
print(candidatos)
#Partidos
for partido in df['sigla_partido']:
    if partido not in partidos:
        partidos.append(partido)
print(partidos)
#UF
for uf in df['sigla_uf']:
    if uf not in ufs:
        ufs.append(uf)
print(ufs)
#Tipo
for tipo in df['tipo_voto']:
    if tipo not in tipos:
        tipos.append(tipo)
print(tipos)
#Cenários
for tipo in df['tipo_voto']:
    if tipo not in tipos:
        tipos.append(tipo)
print(tipos)
```

```python
#Filtrando primeiramente por ano 
df_2022 = df[(df.ano == 2022)]
df_2018 = df[(df.ano == 2018)]
df_2014 = df[(df.ano == 2014)]
df_2010 = df[(df.ano == 2010)]
df_2006 = df[(df.ano == 2006)]
df_2002 = df[(df.ano == 2002)]
```
